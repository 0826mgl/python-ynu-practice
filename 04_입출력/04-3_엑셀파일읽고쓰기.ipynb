```python
# import csv

# import openpyxl
from openpyxl import workbook

wb = workbook()

ws = wb.active

ws.title = '수강생_정보'

ws['A1'] = '철수'
ws['b2'] = '영희'
ws['b3'] = '취미'

wb.save('./수강생리스트.xlsx')

wb.close()
```

```python
from openpyxl import workbook

wb = workbook()
ws = wb.active

ws.title = '수강생정보'

title = ['번호', '이름', '과목']

ws.append(title)

data = [1, '홍길동', '수학']

ws.append(data)

wb.save('./수강생리스트.xlsx')

wb.close()
```
